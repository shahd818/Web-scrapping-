```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
import html.parser
```

```python
full_url='https://books.toscrape.com/index.html'
book_response= requests.get(full_url)
```

```python
book_response
```

```python
text=book_response.text
print(text)
```

```python
soup = BeautifulSoup(text, 'html.parser')
print(soup)
```

```python
urls=[]

for i in range(1,51):
 url = f"https://books.toscrape.com/catalogue/page-{i}.html"
 urls.append(url)
print(urls)
```

```python
books_urls=[]
h3_tags=soup.find_all('h3')
for url in urls:
 for tag in h3_tags:
  a_tags=tag.find('a')
  books_urls.append(a_tags['href'])

books_urls
```

```python
compined_url=[]
for i in books_urls:
 book_url= 'https://books.toscrape.com/' + i
 compined_url.append(book_url)
print(compined_url)
```

```python
books_names=[]
books_prices=[]
books_rating=[]
books_descriptions=[]
books_UPCs=[]
books_availability=[]

for link in compined_url:
 book_response= requests.get(link)
 text1=book_response.text
 soup=BeautifulSoup(text1,'html.parser')

 name=soup.find('h1').text
 books_names.append(name)

 price=soup.find('p', class_='price_color').text.strip('Â£')
 books_prices.append(price)

 rates = soup.find('p', class_='star-rating')
 books_rating.append(rates['class'][1])

 description=soup.find('div' ,id='product_description').find_next('p').text
 books_descriptions.append(description)

 UPC=soup.find('td').text
 books_UPCs.append(UPC)

 available=soup.find_all('td')[-2].text[:8]
 books_availability.append(available)
```

```python
for i in range(len(books_names)):
 print(books_names[i])
 print(books_prices[i])
 print(books_rating[i])
 print(books_descriptions[i])
 print(books_UPCs[i])
 print(books_availability[i])
 print('-----------------------')
```

```python
df=pd.DataFrame({'Book Name':books_names,'Price':books_prices,'Rating':books_rating,'Description':books_descriptions,'UPC':books_UPCs,'Availability':books_availability})
df
```
